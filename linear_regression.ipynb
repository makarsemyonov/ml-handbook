# Линейная регрессия

Пример обучения модели линейной регрессии

---

## Импорт библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

## Загрузка датасета

Чтение данных из CSV, где `x` — признак, `y` — целевая переменная.

```python
data = pd.read_csv("datasets/linear_regression.csv")

X = data[["x"]].values
Y = data["y"].values
```

## График исходных данных

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Данные')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Данные для обучения линейной регрессии')
plt.legend()
plt.grid(True)
plt.show()
```

## Разделение датасета на тестовую и тренировочные выборки

80% датасета это тренировочные данные, 20% - тестовые

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

## Создание модели

```python
model = LinearRegression()
```

## Обучение

```python
model.fit(X_train, Y_train)

w = model.coef_[0] 
b = model.intercept_

print(f"w = {w:.3f}, b = {b:.3f}")
```

## Получение предсказаний на тестовой выборке

```python
Y_test_pred = model.predict(X_test)
```

## Расчет метрик

```python
mse = mean_squared_error(Y_test, Y_test_pred)
mae = mean_absolute_error(Y_test, Y_test_pred)
r2 = r2_score(Y_test, Y_test_pred)

print(f"MSE  = {mse:.3f}")
print(f"MAE  = {mae:.3f}")
print(f"R²   = {r2:.3f}")
```

## Визуализация линии регрессии

```python
plt.figure(figsize=(8, 5))
x_line = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', linewidth=2, label='Линия регрессии')
plt.scatter(X, Y, color='blue', label='Данные')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Данные для обучения линейной регрессии')
plt.legend()
plt.grid(True)
plt.show()
```
